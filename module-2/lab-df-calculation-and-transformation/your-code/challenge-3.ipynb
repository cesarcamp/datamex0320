```python
# your code here
customer_amount=orders.groupby('CustomerID').sum()
```

```python

q_95=customer_amount['amount_spent'].quantile(.95)
q_95
```

```python
q_75=customer_amount['amount_spent'].quantile(.75)
q_75
```

```python
#vip customers
customer_amount.query('amount_spent > 5840.181999999983')
```

```python
#prefered customers
customer_amount.query('amount_spent > 1661.6400000000003 & amount_spent < 5840.181999999983')
```

```python
customer_amount.loc[customer_amount['amount_spent']>= customer_amount['amount_spent'].quantile(0.75), 'Label']='Prefered'
```

```python
customer_amount.loc[customer_amount['amount_spent']>= customer_amount['amount_spent'].quantile(0.95), 'Label']='VIP'
```
