#### Top 10 languages by number of native speakers stored in a Pandas Dataframe

````python
# This is the url you will scrape in this exercise
urll = 'https://en.wikipedia.org/wiki/List_of_languages_by_number_of_native_speakers'
````

````python
#your code
res=requests.get(urll)
data=res.text
soupl=BeautifulSoup(data, 'html.parser')
#soupl
````

````python
tt=soupl.find_all('table')
tlang=tt[0]
#tlang
````

````python
idiomas_pais=[]
for f in tlang.find_all('tr'):
    fila=[e for e in f.find_all('td')]
    
    if len(fila)>0:
        idioma={'language': fila[1].find('a').text.strip(),
              'speakers/millions':(fila[2].text),}
        idiomas_pais.append(idioma)
#idiomas_pais
````

````python
dfidiomas=pd.DataFrame(idiomas_pais)
#dfidiomas
````
