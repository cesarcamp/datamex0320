# WEB SCRAPING

https://en.wikipedia.org/wiki/FIFA_World_Cup

```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
url='https://en.wikipedia.org/wiki/FIFA_World_Cup'
html=requests.get(url).content 
soup=bs(html, 'html.parser')
```

```python
tables=soup.find_all('table')
```

```python
table=tables[2]
table
```

```python
rows=table.find_all('tr')
rows=[r.text.strip().split('\n') for r in rows]
final=[]
for e in rows:
    tmp=[]
    for st in e:
        if st!='':
            tmp.append(st)
    final.append(tmp)
```

```python
final=pd.DataFrame(final)
```

```python
pd.set_option("display.max_rows", None, "display.max_columns", None)
```

```python
final=final.rename(columns=final.iloc[0])
```

```python
final.head()
```

```python
final.columns
```

```python
final.columns = ['Year','Hosts','Venues/Cities','Total_Attendance','Matches','Avg_Attend','Highest_Attend','Venue','Game']
```

```python
final.drop(index=0,inplace=True)
```

```python
final.drop(index=1,inplace=True)
```

```python
final.drop(index=23,inplace=True)
```

```python
final.rename(columns={'Venue':'H.A Venue','Game':'H.A Game'},inplace=True)
```

```python
final.reset_index(inplace = True, drop = True) 
```

```python
final
```

```python
#MERGE DATAFRAMES
mergedDf = fifa.merge(final, left_index=True, right_index=True)
```

```python
mergedDf.head()
```

```python
mergedDf=mergedDf.drop(columns=['MatchesPlayed', 'Attendance','Year_y'])
```

```python
mergedDf
```

```python
mergedDf.loc[[20],'Winner'] = 'France'
mergedDf.loc[[20],'Runners-Up'] ='Croatia'
mergedDf.loc[[20],'Runners-Up'] ='Croatia'
mergedDf.loc[[20],'Third'] ='Belgium'
mergedDf.loc[[20],'Fourth'] ='England'
mergedDf.loc[[20],'GoalsScored'] ='169'
mergedDf.loc[[20],'QualifiedTeams'] ='32'
```

```python
mergedDf.head()
```

```python
mergedDf.rename(columns={'Year_x':'Year'},inplace=True)
```
